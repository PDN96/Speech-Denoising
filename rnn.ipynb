```
M = list()

for i in range(1200):
    M.append(np.greater(signal[i], noise[i]).astype(int))
```

```
model_path = "model.ckpt"

x = tf.placeholder(tf.float32, [10, None, 513])
y = tf.placeholder(tf.float32, [10, None, 513])

n_hidden = 256
w_out = tf.Variable(tf.random_normal([n_hidden, 513], stddev=2/(n_hidden+513), mean=0))
b_out = tf.Variable(tf.zeros([513]))
```

```
lstm_cell = tf.nn.rnn_cell.LSTMCell(n_hidden, initializer=tf.contrib.layers.xavier_initializer())
outputs, _ = tf.nn.dynamic_rnn(lstm_cell, x, dtype=tf.float32)
```

```
weights = tf.expand_dims(tf.ones([10, 1]), 1) * w_out
op = tf.nn.sigmoid(tf.matmul(outputs, weights) + b_out)
```

```
rate = 0.001
mse = tf.reduce_mean(tf.losses.mean_squared_error(op, y))
optimizer = tf.train.AdamOptimizer(learning_rate = rate).minimize(mse)
```

```
init = tf.global_variables_initializer()
saver = tf.train.Saver() 
sess = tf.Session()
sess.run(init)
```

```
nm_epochs = 100

for epoch in range(nm_epochs):
    avg_l = 0
    for i in range(0, 1200, 10):
        batch_x = train[i:i+10] 
        batch_y = M[i:i+10]
        _, l = sess.run([optimizer, mse], feed_dict={x: batch_x, y: batch_y})
        avg_l += l/120
    print("Epoch:",(epoch+1), "\tloss=",(avg_l))
```

```
M_v = []
for i in range(1200):
    M_v.append(np.greater(signal_v[i], noise_v[i]).astype(int))
```

```
s = []
for i in range(1200):
  s.append(librosa.istft(signal_v[i].T , win_length = 1024 , hop_length=512))
```

```
snr = []

for i in range(0, 1200, 10):
    batch_x = train_v[i:i+10] 
    
    with sess.as_default():
      M_hat = op.eval(feed_dict={x:batch_x})
       
    batch_x_complex = X_list_v[i:i+10]
    batch_s = s[i:i+10]
    for j in range(10):
        S_hat = np.multiply(M_hat[j], batch_x_complex[j])
        s_hat = librosa.istft(S_hat.T, win_length=1024, hop_length=512)
        
        t = min(len(s_hat), len(batch_s[j]))
        snr.append(10*np.log10(np.sum(np.square(batch_s[j][:t]))/np.sum(np.square(batch_s[j][:t]-s_hat[:t]))))

print("SNR = ", sum(snr)/1200)
```
